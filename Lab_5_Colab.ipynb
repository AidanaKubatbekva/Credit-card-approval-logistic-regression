## Explanatory Logistic Regression Model

### Prepare Data

### Q3

Write code to complete the following tasks:
- Separate variables into predictors (X) and target (y)
- Dummy encode categorical variables in X (with drop_first = True, dtype = int)
- Map target values: "denied" to label 0 and "approved" to 1 (i.e., positive class)
- Calculate the percentage of 1s (i.e., approved) to check label balance

```python
data_df.columns
```

```python
predictors = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11',
       'V12', 'V13', 'V14', 'V15']
X = data_df[predictors]
y = data_df['Target']

# Dummy encoding
X = pd.get_dummies(X, drop_first = True, dtype = int)

# map labels to 0/1
label_dict = {'Denied': 0, 'Approved': 1}
y = y.map(label_dict)
percentage_of_approved = y.mean() * 100


print(f"Number of predictors in X: {X.shape}")
print(f"Value distribution in y: {y.value_counts()}")
print(f"% of Approved observations: {percentage_of_approved :.2f}%")
```

### Model Development

### Q4

Write code to do the following:

- Fit a logistic regression model using (X,y) with arguments `max_iter = 5000, random_state = 123`
- Print intercept and coefficients

```python

logit_model = LogisticRegression(max_iter= 5000, random_state= 123)
logit_model.fit(X,y)

intercept = logit_model.intercept_
coef_df = pd.DataFrame({"Predictor": X.columns,
                        "Coefficient": logit_model.coef_[0]})


print(f"Intercept: {intercept}")
print(f"Coefficients: :\n{coef_df}")
```

### Q6

Write code to evaluate model performance on training data as follows:
- Accuracy
- Precision/Recall/F1
- Display the actual label, predicted label, and predicted probability of the first 5 observations

```python
# Predict label probability
y_hat_prob = logit_model.predict_proba(X)[:, 1]

# Predict label
y_hat = logit_model.predict(X)

# Create a dataframe with three columns: actual label, predicted label, predicted probability
pred_df = pd.DataFrame({"Actual": y,
                        "Predicted_Label": y_hat,
                        "Predicted_Prob": y_hat_prob})

accuracy = (pred_df. Actual == pred_df.Predicted_Label).mean()

# Print classification report
print(classification_report(y, y_hat))


print(f"Training accuracy: {accuracy :.4f}\n")

print(f"Prediction of first 5 observations: :\n{pred_df.head()}")
```
